## 12 Visualization Hacks That Turn Data Into Stories

### 1. Annotated Timelines That Speak for Themselves
Your plot is worth nothing if someone has to ask “What’s that spike?”. Annotate it.

```python
import matplotlib.pyplot as plt
import pandas as pd

dates = pd.date_range("2024-01-01", periods=30)
values = [i + (i**1.5 % 5) for i in range(30)]

plt.plot(dates, values, marker='o')
plt.annotate(
    'Major Campaign Launch',
    xy=(dates[10], values[10]),
    xytext=(dates[12], values[10]+2),
    arrowprops=dict(facecolor="#63E047", shrink=0.05)
)
plt.show()
```

### 2. Ridgeline Plots for Trend Overlap (Not Just for Hipsters)
When multiple distributions overlap over time, ridgeline plots tell the story beautifully.

```python
# %pip install joypy
```

```python
import joypy
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'value': np.random.randn(2000),
    'month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr'], 2000)
})

joypy.joyplot(df, by='month', column='value', colormap=plt.cm.coolwarm)
plt.show()
```

### 3. Color Gradients That Tell Magnitude, Not Just Mood
Ditch default blues. Make your color actually mean something.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 10)
plt.imshow(data, cmap='viridis')
plt.colorbar(label='Revenue in $M')
plt.show()
```

### 4. Small Multiples for Pattern Hunting
One plot per category, arranged neatly. Think comic strip for your dataset.

```python
import seaborn as sns
import pandas as pd

penguins = sns.load_dataset("penguins")
sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm",
    col="species", hue="sex", style="sex"
)
plt.show()
```

### 5. Motion Plots (Animated Stories)
Sometimes the story is in the change, not the snapshot.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i/10))
    return line,

ani = animation.FuncAnimation(fig, animate, frames=100, interval=50)
plt.show()
```

### 6. Slope Charts for Before-and-After Truth Bombs
Bar charts hide the relationship. Slope charts reveal direction instantly.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'category': ['A', 'B', 'C'],
    '2023': [4, 6, 7],
    '2024': [8, 5, 9]
})

for _, row in df.iterrows():
    plt.plot([1, 2], [row['2023'], row['2024']], marker='o')
    plt.text(1, row['2023'], row['category'])
plt.xticks([1, 2], ['2023', '2024'])
plt.show()
```

### 7. Sankey Diagrams That Explain Complex Flows Instantly
Perfect for showing how inputs become outputs without drowning in arrows.

```python
import plotly.graph_objects as go

fig = go.Figure(go.Sankey(
    node=dict(label=["A", "B", "C", "D"]),
    link=dict(source=[0, 1, 0], target=[2, 3, 3], value=[8, 4, 2])
))
fig.show()
```

### 8. Highlight-Only-the-Important Bars
Why plot everything the same color when you only care about one bar?

```python
import matplotlib.pyplot as plt

data = [5, 7, 3, 8, 6]
bars = plt.bar(range(len(data)), data)
bars[2].set_color('red')
plt.show()
```

### 9. Bump Charts for Ranking Over Time
If the plot looks like spaghetti, you’re doing it wrong.

```python
import pandas as pd
import plotly.express as px

df = pd.DataFrame({
    'year': [2020, 2020, 2021, 2021],
    'team': ['A', 'B', 'A', 'B'],
    'rank': [2, 1, 1, 2]
})
px.line(df, x='year', y='rank', color='team', markers=True).show()
```

### 10. Histogram + KDE Combo
Bars for the literal data. Curve for the trend. Instant credibility.

```python
import seaborn as sns
import numpy as np

sns.histplot(np.random.randn(500), kde=True)
plt.show()
```

### 12. Layered Context with Dual Axes (Carefully)
One chart, two scales — if your data really needs it.

```python
import matplotlib.pyplot as plt

x = range(10)
y1 = [i for i in x]

y2 = [i**2 for i in x]
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

plt.show()
```
