Since PyWaffle is built on Matplotlib, make sure you have Matplotlib installed as well.

#### 2. Your First Waffle Chart
Letâ€™s create a simple waffle chart showing pet preferences:

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

data = {'Group A': 100, 'Group B': 200, 'Group C': 300}
fig = plt.figure(
    FigureClass=Waffle,
    rows=20,
    values=data,
    colors=["#308df0ff", "#063e97ff", "#021459ff"],
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}
)
plt.show()
```

#### 3. Adding Icons for More Fun
PyWaffle supports FontAwesome icons, which makes waffle charts more engaging:

https://fontawesome.com/

```python
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values={'Coffee': 40, 'Tea': 60},
    icons='coffee',    # use FontAwesome icons
    icon_size=18,
    colors=["#6f4e37", "#c5a880"],
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}
)
plt.show()
```

```python
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values={'Headphone on': 80, 'Headphone off': 30},
    icons='headphones',    # use FontAwesome icons
    icon_size=18,
    colors=["#f5079e", "#f5d509"],
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}
)
plt.show()
```

#### 4. Multiple Subplots
You can also create multiple waffle charts side by side for comparison:

```python
fig = plt.figure(
    FigureClass=Waffle,
    plots={
        211: {
            'values': {'Male': 55, 'Female': 45},
            'title': {'label': 'Gender Distribution', 'loc': 'center'}
        },
        212: {
            'values': {'Yes': 70, 'No': 30},
            'title': {'label': 'Survey Response', 'loc': 'center'}
        }
    },
    rows=5,
    colors=["#4c72b0", "#dd8452"]
)
plt.show()
```

#### 5. Custom Layouts and Styling
You can fine-tune waffle charts to fit your presentation style:

Adjust rows to change chart resolution.
Change icon_size for larger or smaller icons.
Add annotations or percentages with matplotlib.text.
Use different color palettes for a professional look.
Example with percentages:

```python
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values={'A': 25, 'B': 75},
    colors=["#8da0cb", "#fc8d62"],
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    labels=["A (25%)", "B (75%)"]
)
plt.show()
```
